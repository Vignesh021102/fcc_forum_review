```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("fcc-forum-pageviews.csv")
```

```python
data = data[(data.value > np.percentile(data.value,2.5))&(data.value < np.percentile(data.value,97.5))]
```

```python
lin = np.linspace(0,data.date.shape[0]-1,8).round()
date = np.char.ljust(list(data.iloc[lin].date.values),7)
```

```python
fig = plt.figure(figsize=(10,5))
ax = plt.axes()
ax.plot(np.arange(0,data.date.shape[0]),data.value.values,'r',linewidth=0.5)
ax.set(box_aspect = 0.4,title="Daily freeCodeCamp Froum Page Views 5/2016-12/2019",xlabel="Date",ylabel="Page Views")
plt.xticks(lin,date)
ax.tick_params(labelsize= 7)
ax.axis('tight')
plt.show()
```

```python
fig.savefig('plot_graph.jpeg',dpi = 200)
```

```python
data.date = np.char.ljust(list(data.date.values),7)
```

```python
data["month"] = np.char.partition(list(data.date.values),'-')[0:,2]
data["year"] = np.char.partition(list(data.date.values),'-')[0:,0]
```

```python
monthLabel = ["Jan","Feb","Mar","Apr","May","Jun","Jul","Aug","Sepr","Oct","Nov","Dec"]
yearLabel = ["2016","2017","2018","2019"]
```

```python
br = np.arange(0,4*15,15)
fig = plt.figure(figsize=(7,7))
ax = plt.axes()

for i in data.groupby("month"):
    j = data[data.month == i[0]].groupby("year").mean()
    if "2016" not in j.index:
        j.loc["2016"] = 0
        j = j.sort_index()
    br=br+1
    ax.bar(br,j.values.reshape(4,),width=1)
plt.legend(monthLabel,title="Month")
plt.xlabel("Years")
plt.ylabel("Average page viwes")
plt.xticks(br-6,yearLabel)
plt.show()
fig.savefig("bar_chart.jpeg",dpi=500)
```
